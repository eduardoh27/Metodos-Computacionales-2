```python
import numpy as np 
import matplotlib.pyplot as plt

def fourier_series(n, L, x):
    fourier_sum = np.zeros(len(x))
    for i in range(1, n+1):
        fourier_sum += (2/(i*np.pi)) * (1 - (-1)**i) * np.sin(i*np.pi*x/L)
    return fourier_sum

x_plot = np.linspace(-6, 6, 1000)
y_plot = fourier_series(300, 1, x_plot)

plt.plot(x_plot, y_plot)
plt.show()
```
