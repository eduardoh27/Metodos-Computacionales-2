En este problema vamos a estudiar la tranformada discreta de Fourier para el an치lisis de imagenes. Considere la imagen "blurredcity.png", de la cual tomamos solo una secci칩n de la imagen. Para poder leer la imagen este notebook debe estar en la carpeta Datafiles del github del curso.

```python
# Lectura de la imagen de circulos, no modificar
image_city = plt.imread('blurredcity.png').astype(float)
circles_image = image_city[300:350, 170:220]

plt.imshow(circles_image)
```

a) Implemente la transformada discreta de Fourier en 2D dada por, 

$$
  F[k, l] = \frac{1}{\sqrt{MN}}\sum_{m=0}^{M-1}\sum_{n=0}^{N-1}f[m, n]\exp{\Big[-i2\pi\big(\frac{km}{M}+ \frac{ln}{N}\big)\Big]}
$$

La cual recibe una imagen de dimensi칩n (m, n), y retorna una matrix de dimensi칩n (m, n) que corresponde al espacio de frecuencias.

```python
def DFT_lenta_2D(image_param):
    N = image_param.shape[0]
    M = image_param.shape[1]
    frequency_matrix = np.zeros((N, M)).astype(complex)

    
    for k in range(M):
        for l in range(N):
            sumatoria = 0
            for m in range(M):
                for n in range(N):
                    sumatoria += (image_param[m][n][0] * np.exp(-1j*2*np.pi*(k*m/M + l*n/N)))      
            frequency_matrix[k][l] = sumatoria/np.sqrt(M*N)
            
    return frequency_matrix

frequency_image = DFT_lenta_2D(circles_image)
```

$$

La cual transforma una imagen de (m, n) del espacio de frecuencias, a una matrix (m, n) del espacio de pixeles de la imagen.

```python
def DFT_lenta_inversa_2D(image_param):
    N = image_param.shape[0]
    M = image_param.shape[1]
    time_matrix = np.zeros((N, M)).astype(complex)
    
    
    for m in range(M):
        for n in range(N):
            sumatoria = 0
            for k in range(M):
                for l in range(N):
                    sumatoria += (image_param[k][l] * np.exp(1j*2*np.pi*(k*m/M + l*n/N)))      
            time_matrix[m][n] = sumatoria/np.sqrt(M*N)
            
    

    return time_matrix

new_image = DFT_lenta_inversa_2D(frequency_image)
```
