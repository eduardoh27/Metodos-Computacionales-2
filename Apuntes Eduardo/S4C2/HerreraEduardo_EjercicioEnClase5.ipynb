## Problema 

Una función de onda viajera $\psi(x, t) = e^{i(kx - \omega t)}$ compleja, se puede interpretar como dos ondas simultaneas una con componente real y otra con componente imaginaria, 

$$
e^{i(kx - \omega t)} = \cos{(kx - \omega t)} + i \sin{(kx - \omega t)}
$$

a) Tomando $t=0$, gráfique con una figura y dos ejes (uno para la real otro para la imaginaria), las dos ondas para $k = 0.5$, y para valores de x entre $-20$ y $20$. Puede guiarse del siguiente [notebook](https://github.com/diegour1/MetodosComputacionales2/blob/main/Notebooks/07%20-%20Pyplot_and_animation.ipynb).

```python
# su codigo aqui
k = 0.5
x = np.linspace(-20, 20, 100)
y1 = np.cos(k*x)
y2 = np.sin(k*x)

fig, ax = plt.subplots(2)
ax[0].plot(x, y1)
ax[1].plot(x, y2)
```

b) Realice una simulación de las dos ondas viajeras para $t$ entre $0$ y $10$ con 100 puntos con $\omega = 1$, las dos ondas deben desplazarse de forma simultanea. La simulación debe aparecer en el notebook, no deberia aparecer como un archivo separado.

```python
# su codigo aqui
k = 0.5
x = np.linspace(-20, 20, 100)
t = np.linspace(0, 10, 100)
w = 1

matriz_y1 = []
matriz_y2 = []


for i in range(len(t)):
    y1 = np.cos(k*x - w*t[i])
    y2 = np.sin(k*x - w*t[i])
    matriz_y1.append(y1)
    matriz_y2.append(y2)

# primera fila tiempo = 0, toda la onda.
# tiempo = 1 calcula toda la onda.

fig, ax = plt.subplots(2)
fig.figsize=(6, 4)
ax[0].set_xlim([-20, 20])
ax[0].set_ylim([-1, 1])
ax[1].set_xlim([-20, 20])
ax[1].set_ylim([-1, 1])


def update(i):
    ax[0].clear()
    ax[1].clear()
    ax[0].plot(x, matriz_y1[i], "blue")
    ax[1].plot(x, matriz_y2[i], "red")

ani.FuncAnimation(fig, func = update, frames=range(30))
     
```
